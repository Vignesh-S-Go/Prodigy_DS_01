# âœ… Import libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

Load the dataset

```python

uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)
```

Data Preview

```python
data.head()
```

 âœ… Convert population column to numeric

```python
data['2022 Population'] = pd.to_numeric(data['2022 Population'], errors='coerce')
```

âœ… Sort and select top 10 countries

```python
top10 = data.sort_values(by='2022 Population', ascending=False).head(10)
```

Plot Barchat

```python
plt.figure(figsize=(12, 7))
sns.barplot(
    x=top10['2022 Population'],
    y=top10['Country/Territory'],
    palette='viridis'
)

plt.title('Top 10 Most Populated Countries in 2022', fontsize=16)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# âœ… Plot histogram

```python

plt.figure(figsize=(10, 6))
sns.histplot(data['2022 Population'], bins=30, kde=True, color='skyblue')

plt.title('Population Distribution Across Countries (2022)', fontsize=16)
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
